## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
#agrupamento
segmento = data.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda',
                                                                                      ascending = False)
```

```python
#função para que os valores não fiquem em notação cientifica
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
# Gáfico

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(segmento['Valor_Venda'], 
        labels = segmento['Segmento'],
        autopct = autopct_format(segmento['Valor_Venda']),
        startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(segmento['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

## Pergunta de Negócio 8:

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
data['Valor_Venda_Com_Desconto'] = data['Valor_Venda'] - (data['Valor_Venda'] * data['Desconto'])
```

```python
#Filtrando o valor de vendas 15% antes do desconto
sales_before_discount = data.loc[data['Desconto']==0.15, 'Valor_Venda']
```

```python
#Filtrando o valor das vendas depois do desconto de 15%
sales_after_discount = data.loc[data['Desconto'] ==0.15, 'Valor_Venda_Com_Desconto']
```

```python
#Calculando a média antes do desconto
sales_before_discount.mean()
```

```python
#Calculando a média depois do desconto
sales_after_discount.mean()
```

## Pergunta de Negócio 9:

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
data['Mes'] = data['Data_Pedido'].dt.month
```

```python
month_year_segment = data.groupby(['Ano','Mes','Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
years = month_year_segment.index.get_level_values(0)
months = month_year_segment.index.get_level_values(1)
segment = month_year_segment.index.get_level_values(2)
```

```python
# Plot
plt.figure(figsize = (12, 6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                   data = month_year_segment, 
                   y = 'mean', 
                   x = months,
                   hue = segment,  #preenchimento
                   col = years, #colunas
                   col_wrap = 4) 
plt.show()
```
